```python
import pandas as pd
import matplotlib.pyplot as plt


before_complete = pd.read_csv('statistics/fitness_statistics_before_complete.csv')
after_complete = pd.read_csv('statistics/fitness_statistics_after_complete.csv')


before_complete['Phase'] = 'Before Complete'
after_complete['Phase'] = 'After Complete'


combined_data = pd.concat([before_complete, after_complete], ignore_index=True)


plt.figure(figsize=(12, 6))

for phase, data in combined_data.groupby('Phase'):
    plt.plot(data['Generation'], data['Average Fitness'], label=phase)

plt.title('Average Fitness Over Generations')
plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(12, 6))
for phase, data in combined_data.groupby('Phase'):
    plt.plot(data['Generation'], data['Std Fitness'], label=f'{phase} (Std Fitness)')

plt.title('Standard Deviation of Fitness Over Generations')
plt.xlabel('Generation')
plt.ylabel('Standard Deviation of Fitness')
plt.legend()
plt.grid()
plt.show()
```

```python

correlation_before = before_complete[['Generation', 'Average Fitness']].corr().iloc[0, 1]
correlation_after = after_complete[['Generation', 'Average Fitness']].corr().iloc[0, 1]

print(f"Correlation between generations and average fitness (Before Complete): {correlation_before:.2f}")
print(f"Correlation between generations and average fitness (After Complete): {correlation_after:.2f}")
```

The correlation values indicate a strong negative relationship between generations and average fitness for both the Before Complete and After Complete phases. This means that as the generation increases, the average fitness decreases.
